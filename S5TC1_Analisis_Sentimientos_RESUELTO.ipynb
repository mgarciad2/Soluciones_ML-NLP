### Punto 1 - Uso de CountVectorizer

En la celda 1 creen un modelo de random forest con la libreria sklearn que prediga el sentimiento de los tweets usando los set de entrenamiento y test definidos anteriormente. Usen la función **CountVectorizer** y presenten el desempeño del modelo con la métrica del acurracy.

Recuerden que el preprocesamiento que se haga sobre los datos de entrenamiento  (*.fit_transform()*) deben ser aplicado al set de test (*.transform()*).

```python
# Celda 1
# Creación de matrices de documentos usando CountVectorizer con todo en vocabulario en minúsculas.
vect_1 = CountVectorizer() 
X_train_1 = vect_1.fit_transform(X_train) # Preprocesamiento en train
X_test_1 = vect_1.transform(X_test) # Preprocesamiento en test
```

En la celda anterior **generamos la tokenización y el vocabulario a partir de los tweets que están en el conjunto de Train. Ademas, realizamos la misma transformación a los datos de Test**. 

En la siguiente celda **vamos a calibrar y entrenar el modelo con GridSearchCV**. Hay que aclarar que como se vio en las gráfica de pastel realizadas en la sección de preprocesamiento y arreglo de los datos **hay un desbalanceo notable en las 3 clases/sentimientos de los Tweets, por lo tanto para tratar con ella lo implementamos directamente en  los parámetros del Random Forest Classifier, colocando dos parametros para realizar el balanceo: *oob=True* y el *class_weight="balanced"*, como se puede ver a continuación.**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_jobs=-1,max_features= 'sqrt' ,n_estimators=50, oob_score = True,class_weight = "balanced") 

param_grid = { 
    'n_estimators': [200, 400],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8]
}
CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 5)
CV_rfc.fit(X_train_1, y_train)
print(CV_rfc.best_params_)
```

A partir de la calibracióncon GridSearchCV usamos el mejor modelo entrenado para **predecir sobre el conjunto de Test, tanto el tipo de Tweet como la propabilidad en cada clase**:

```python
y_pred_1=CV_rfc.best_estimator_.predict(X_test_1)
y_pred_1p=CV_rfc.best_estimator_.predict_proba(X_test_1)
```

```python
unique, counts = np.unique(y_pred_1,return_counts=True)
print(np.asarray((unique,counts)).T)
```

Ya con las predicciones hechas **calculamos un accuracy global, aunque no se considera una buena medida para un problema multiclase como este**.

```python
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, auc
acc_rf = accuracy_score(y_test, y_pred_1) # Accuracy en test
print('Accuracy evaluado respecto a la muestra de test: ', acc_rf)
```

También presentamos **mediciones más adecuadas directamente sobre cada clase/sentimiento del Tweet (como son precision, recall y f1-score):**

```python
from sklearn.metrics import classification_report
target_names = ['class 0', 'class 1', 'class 2']
print(classification_report(y_test, y_pred_1, target_names=target_names))
```

Presentamos también de forma visual **la matriz de confusión** con las diferentes clases reales y las predichas. De la gráfica de la matriz de confusión podemos ver la **proporción de errores cometidos al ojear la cantidad de elementos que están fuera de la diagonal**, que aunque son bastantes no son demasiados si los comparamos con los valores que están sobre la diagonal.

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
conf_mat = confusion_matrix(y_test, y_pred_1)
fig, ax = plt.subplots(figsize=(4,4))
sns.heatmap(conf_mat, annot=True, fmt='d',
            xticklabels=target_names, yticklabels=target_names)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
roc_auc_score(y_test, y_pred_1p, multi_class='ovr')
```

Por último para completar nuestro análisis vamos a dar **el AUC y la curva ROC** para los cada una de las clases de sentimientos de los Tweets.

```python
# roc curve for classes

fpr = {}
tpr = {}
thresh = {}
auc_roc = {}

clases = [-1,0,1]


for i, value in enumerate(clases):    
    fpr[i], tpr[i], thresh[i] = roc_curve(y_test, y_pred_1p[:,i], pos_label=value)
    auc_roc[i] = auc(fpr[i], tpr[i])
    
# plotting    
plt.plot(fpr[0], tpr[0], linestyle='--',color='orange', label='Class 0 vs Rest - '+str(round(auc_roc[0],2)))
plt.plot(fpr[1], tpr[1], linestyle='--',color='green', label='Class 1 vs Rest - '+str(round(auc_roc[1],2)))
plt.plot(fpr[2], tpr[2], linestyle='--',color='blue', label='Class 2 vs Rest - '+str(round(auc_roc[2],2)))
plt.title('Multiclass ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive rate')
plt.legend(loc='best')
plt.savefig('Multiclass ROC',dpi=300); 
```

En la gráfica podemos ver el **comportamiento de la curva ROC para cada una de las clases comparadas con el resto y el valor del AUC en cada una de ellas**. Estos valores están dentro de un rango adecuado.

```python
auc_roc
```

Para el proceso de lematización creamos la **función que va a dividir los datos de entrenamiento en lemas**:

```python
# Definición de la función que tenga como parámetro texto y devuelva una lista de lemas
def split_into_lemmas(text):
    text = text.lower()
    words = text.split()
    return [wordnet_lemmatizer.lemmatize(word) for word in words]
```

```python
# Creación de matrices de documentos usando CountVectorizer, usando el parámetro 'split_into_lemmas'
vect_lemas = CountVectorizer(analyzer=split_into_lemmas)
```

Generamos la **tokenización con lemas tanto para el conjunto de Train como para test**, a los dos se les realiza la misma transformación:

```python
# Creación de matrices de documentos usando CountVectorizer sin stopwords
X_train_3 = vect_lemas.fit_transform(X_train) # Preprocesamiento en train
X_test_3 = vect_lemas.transform(X_test) # Preprocesamiento en test
```

```python
rfc = RandomForestClassifier(n_jobs=-1,max_features= 'sqrt' ,n_estimators=50, oob_score = True,class_weight = "balanced") 

param_grid = { 
    'n_estimators': [200, 400],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8]
}
CV_rfc_l = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 5)
CV_rfc_l.fit(X_train_3, y_train)
print(CV_rfc_l.best_params_)
```
